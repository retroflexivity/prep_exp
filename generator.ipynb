```python
import re
```

```python
def st_subst(s: str, mod: str, pron: str, dupl: bool, pos: str):
    if dupl:
        pron = 'у ' + pron
        mod = 'у ' + mod
    elif pos == 'L':
        mod = 'у ' + mod
    else:
        pron = 'у ' + pron
        
    s = re.sub(f'MOD_{pos}', mod, s)
    s = re.sub('MOD_.', '', s)
    s = re.sub(f'PRON_{pos}', pron, s)
    s = re.sub('PRON_.', '', s)
    s = re.sub('MOD', mod, s)
    s = re.sub('PRON', pron, s)
    
    s = s.replace('  ', ' ').strip()
    s = s[0].upper() + s[1:]
    
    return s
```

```python
stimuli_all = []
dupls = (True, False)
poss = ('R', 'L', 'E')

for dupl in dupls:
    for pos in poss:
        for r in stimuli.apply(lambda row: st_subst(row['Sentence'], row['mod'], row['pron'], dupl, pos), axis=1):
            stimuli_all.append({'stim': r, 'dupl': dupl, 'pos': pos})
            
stimuli_all = pd.DataFrame(stimuli_all)
```

```python
import numpy as np
```

```python
stimuli_all['modifier'] = np.tile(stimuli['modifier'], 6)
stimuli_all['syntactic_position'] = np.tile(stimuli['position'], 6)
```

```python
stimuli_all
```

```python
stimuli_all = stimuli_all.drop(stimuli_all[(stimuli_all['syntactic_position'] == 'low') & (stimuli_all['pos'] == 'L')].index)
```

```python
stimuli_all['group'] = (1, 2) * (len(stimuli_all) // 4) + (2, 1) * (len(stimuli_all) // 4)
```

```python
stimuli_all.to_csv('stimuli_all.csv', index=False)
```
